```python
import matplotlib.pyplot as plt
```

```python
sequential = [960900, 1898300, 864600]
parallel = [200100, 311800, 184600]
gpu = {
    "1b4t": [40, 60, 43],
    "1b8t": [66, 55, 62],
    "1b16t": [64, 58, 41],
    "1b32t": [79, 60, 41],
    "2b4t": [64, 57, 59],
    "2b8t": [65, 59, 64],
}
```

```python
# speedups as to sequential
speedups_sequential = {}
for run in gpu:
    entry = gpu[run]
    final_entry = []
    for i in range(3):
        final_entry.append(sequential[i] / entry[i])
    speedups_sequential[run] = final_entry
```

```python
# speedups as to parallel
speedups_parallel = {}
for run in gpu:
    entry = gpu[run]
    final_entry = []
    for i in range(3):
        final_entry.append(parallel[i] / entry[i])
    speedups_parallel[run] = final_entry
```

```python
entries = [1, 2, 3, 4, 5, 6]
xauto = [int(x) - 0.2 for x in entries]
xdelaunay = [int(x) for x in entries]
xgb = [int(x) + 0.2 for x in entries]
yauto = [speedups_sequential["1b4t"][0], speedups_sequential["1b8t"][0], speedups_sequential["1b16t"][0], speedups_sequential["1b32t"][0], speedups_sequential["2b4t"][0], speedups_sequential["2b8t"][0]]
ydelaunay = [speedups_sequential["1b4t"][1], speedups_sequential["1b8t"][1], speedups_sequential["1b16t"][1], speedups_sequential["1b32t"][1], speedups_sequential["2b4t"][1], speedups_sequential["2b8t"][1]]
ygb = [speedups_sequential["1b4t"][2], speedups_sequential["1b8t"][2], speedups_sequential["1b16t"][2], speedups_sequential["1b32t"][2], speedups_sequential["2b4t"][2], speedups_sequential["2b8t"][2]]
width = 0.2
plt.figure(figsize=(12, 8))
plt.bar(xauto, yauto, width)
plt.bar(xdelaunay, ydelaunay, width)
plt.bar(xgb, ygb, width)
plt.legend(["auto", "delaunay_n22", "great-britain_osm"])
plt.xlabel("CUDA Configuration (1: 1B4T, 2: 1B8T, 3: 1B16T, 4: 1B32T, 5: 2B4T, 6: 2B8T)")
plt.ylabel("Speedup compared to sequential version")
plt.title("Speedup (to sequential)")
plt.savefig("figs/speedup_seq.png")
```
